```python
df_sorted = df.sort_values (by='Date')

df_sorted.set_index('Date', inplace=True)

df_sorted
```

```python
df_sorted = df_sorted.drop('Competition', axis = 1)

df_sorted
```

```python
plt.figure(figsize=(20, 10))

# Plot with marker and linestyle
plt.plot(df_sorted.index, df_sorted['Throw'], marker='o', linestyle='-', label='Throws') 

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Annotate each point on the line plot
for x, y in zip(df_sorted.index, df_sorted['Throw']): 
    plt.annotate(
        f'{y}', 
        (x, y), 
        textcoords="offset points", 
        xytext=(0,5), 
        ha='center'
    )

plt.title('Line Plot with Annotations')

plt.xlabel('X-axis Label')

plt.ylabel('Y-axis Label')

plt.legend()

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from datetime import datetime

df_sorted.index = pd.to_datetime(df_sorted.index)

# Map dates to ordinal values for regression
df_sorted['Date_ordinal'] = df_sorted.index.map(datetime.toordinal)

# Prepare feature and target variables
X = df_sorted['Date_ordinal'].values.reshape(-1, 1)
y = df_sorted['Throw']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict using the model
df_sorted['Predicted_Throw'] = model.predict(X)

# Plotting
plt.figure(figsize=(10, 6))

# Plot actual throw values
plt.plot(df_sorted.index, df_sorted['Throw'], label='Actual Throw', marker='o')

# Plot predicted throw values
plt.plot(df_sorted.index, df_sorted['Predicted_Throw'], label='Predicted Throw', linestyle='--')

plt.xlabel('Date')
plt.ylabel('Throw')
plt.title('Linear Regression on Throw Data Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_sorted already exists with 'Actual_Throw' and 'Predicted_Throw'
# Define bins for categorizing the throw distances
bins = [0, 85, 87, np.inf]
labels = ['Short', 'Medium', 'Long']

# Create new columns for categorized throw distances
df_sorted['Actual_Category'] = pd.cut(df_sorted['Throw'], bins=bins, labels=labels)
df_sorted['Predicted_Category'] = pd.cut(df_sorted['Predicted_Throw'], bins=bins, labels=labels)

# Display the first few rows to see the categorized values
print(df_sorted[['Throw', 'Predicted_Throw', 'Actual_Category', 'Predicted_Category']].head())
# Generate the confusion matrix
conf_matrix = confusion_matrix(df_sorted['Actual_Category'], df_sorted['Predicted_Category'], labels=labels)

# Plotting the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Category')
plt.ylabel('Actual Category')
plt.show()

```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import precision_recall_curve, f1_score
import matplotlib.pyplot as plt

# Assuming df_sorted already exists with 'Actual_Throw' and 'Predicted_Throw'

# Define the threshold for binary classification (e.g., above or below 87)
threshold = 87

# Convert actual and predicted throws to binary categories based on the threshold
df_sorted['Actual_Binary'] = (df_sorted['Throw'] > threshold).astype(int)
df_sorted['Predicted_Throw_Prob'] = (df_sorted['Predicted_Throw'] > threshold).astype(int)

# Get prediction probabilities
y_true = df_sorted['Actual_Binary']
y_scores = df_sorted['Predicted_Throw']

# Calculate precision, recall, and thresholds
precisions, recalls, thresholds = precision_recall_curve(y_true, y_scores)

# Calculate F1 scores for each threshold
f1_scores = [f1_score(y_true, y_scores > t) for t in thresholds]

# Plotting F1 Score vs. Threshold Curve
plt.figure(figsize=(10, 6))
plt.plot(thresholds, f1_scores, marker='o', linestyle='-', color='b')
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs. Threshold Curve')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Prepare feature and target variables (same as before)
X = df_sorted['Date_ordinal'].values.reshape(-1, 1)
y = df_sorted['Throw']

# Create and fit the Random Forest regression model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# Predict using the Random Forest model
df_sorted['Predicted_Throw_RF'] = rf_model.predict(X)

# Calculate performance metrics
mae_rf = mean_absolute_error(y, df_sorted['Predicted_Throw_RF'])
mse_rf = mean_squared_error(y, df_sorted['Predicted_Throw_RF'])
print(f"Random Forest MAE: {mae_rf}")
print(f"Random Forest MSE: {mse_rf}")

# Plotting
plt.figure(figsize=(10, 6))

# Plot actual throw values
plt.plot(df_sorted.index, df_sorted['Throw'], label='Actual Throw', marker='o')

# Plot predicted throw values (Random Forest)
plt.plot(df_sorted.index, df_sorted['Predicted_Throw_RF'], label='Predicted Throw (RF)', linestyle='--')

plt.xlabel('Date')
plt.ylabel('Throw')
plt.title('Random Forest Regression on Throw Data Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Prepare feature and target variables (same as before)
X = df_sorted['Date_ordinal'].values.reshape(-1, 1)
y = df_sorted['Throw'].values.reshape(-1, 1)

# Feature scaling
scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y).ravel()

# Create and fit the SVR model
svr_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_model.fit(X_scaled, y_scaled)

# Predict using the SVR model
y_pred_scaled = svr_model.predict(X_scaled)
df_sorted['Predicted_Throw_SVR'] = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))

# Calculate performance metrics
mae_svr = mean_absolute_error(y, df_sorted['Predicted_Throw_SVR'])
mse_svr = mean_squared_error(y, df_sorted['Predicted_Throw_SVR'])
print(f"SVR MAE: {mae_svr}")
print(f"SVR MSE: {mse_svr}")

# Plotting
plt.figure(figsize=(10, 6))

# Plot actual throw values
plt.plot(df_sorted.index, df_sorted['Throw'], label='Actual Throw', marker='o')

# Plot predicted throw values (SVR)
plt.plot(df_sorted.index, df_sorted['Predicted_Throw_SVR'], label='Predicted Throw (SVR)', linestyle='--')

plt.xlabel('Date')
plt.ylabel('Throw')
plt.title('Support Vector Regression on Throw Data Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
